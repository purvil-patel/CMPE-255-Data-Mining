```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4, random_state=0)
gmm.fit(X)
```

```python
labels = gmm.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.show()
```

```python
import numpy as np
from sklearn.metrics import silhouette_score

# Assuming X is your dataset and labels are the cluster labels from GMM
# Calculate the silhouette score
silhouette_avg = silhouette_score(X, labels)
print(f"Silhouette Score: {silhouette_avg:.2f}")

# If you have true labels (y_true) from make_blobs, you can compare them
# This part is optional and only applicable if true labels are available
comparison = np.mean(labels == y_true)
print(f"Accuracy compared to true labels: {comparison:.2f}")

# Visualize the centroids and shapes of each cluster
# Extract cluster centers and covariances
centers = gmm.means_
covariances = gmm.covariances_

# Plotting the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
# Plot the centers
for center in centers:
    plt.scatter(center[0], center[1], c='black', s=200, alpha=0.5)
plt.title("GMM Clustering Results with Cluster Centers")
plt.show()
```
